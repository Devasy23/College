```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("images.jpeg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
s=[]
for i in img:
    for j in i:
        s.append(j)
s.sort()
s
```

```python
unq=[]
for i in s:
    if i not in unq:
        unq.append(i)
unq.sort()
unq
```

```python
freq={}
n = len(s)
x=256
for i in range(x):
    count=0
    for j in range(n):
        if (i == s[j]):
            count += 1
    freq[i]=count
freq
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(range(len(freq)), list(freq.values()), alpha=0.7, color='gray',align='center')
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x
plt.grid()
plt.show()
```

```python
cm={}
n = len(s)
x=256
for i in range(x):
    count=0
    for j in range(n):
        if (i >= s[j]):
            count += 1
    cm[i]=count
cm
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(range(len(cm)), list(cm.values()), color='gray',align='center',alpha=0.7)
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x
plt.grid()
plt.show()
```
