```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
sqimage = cv2.imread('image.jpg')
```

```python
def scale_image(sqimage, powerof2=2):
  for power in range(powerof2):
    arr = np.zeros((2*sqimage.shape[0], 2*sqimage.shape[1], 3), dtype='int')
    print(arr.shape)
    for k in range(3):
        for j in range(sqimage.shape[1]):
            for i in range(sqimage.shape[0]):
                try:

                    arr[2*i][2*j][k] = sqimage[i][j][k]
                    arr[2*i+1][2*j][k] = sqimage[i][j][k]
                    arr[2*i][2*j+1][k] = sqimage[i][j][k]
                    arr[2*i+1][2*j+1][k] = sqimage[i][j][k]
                except:
                    print(i, j, k)
    sqimage = arr


  return arr

```

```python
arr = scale_image(sqimage, 3)
plt.imshow(arr)
arr = np.array(arr, dtype='uint8')
```

```python
plt.imsave('resizedimg.jpg', arr)
```
