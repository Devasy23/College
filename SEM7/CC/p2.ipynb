# Recursive descent parsing

```python
# print("THIS EXAMPLE IS TAKEN FROM PRACTICAL LIST, 3")
print("S->AS'\nS'->+AS'/a\nA->FA'\nA'->*FA'/a\nF->(S)/i\n")
global n

n = list(input("Enter the string\n"))



print("Entered String is : ", n)
global i
i = 0

def match(a):
    global n
    global i
    if i >= len(n):
        return False
    elif n[i] == a:
        i += 1
        return True
    else:
        return False

def F():
    if match("("):
        if S():
            if match(")"):
                return True
            else:
                return False
        else:
            return False
    elif match("i"):
        return True
    else:
        return False

def Ax():
    if match("*"):
        if F():
            if Ax():
                return True
            else:
                return False
        else:
            return False
    else:
        return True

def A():
    if F():
        if Ax():
            return True
        else:
            return False
    else:
        return False

def Sx():
    if match("+"):
        if A():
            if Sx():
                return True
            else:
                return False
        else:
            return False
    else:
        return True

def S():
    if A():
        if Sx():
            return True
        else:
            return False
    else:
        return False

if S():
    if i == len(n):
        print("String is accepted")
    else:
        print("String is not accepted")
else:
    print("string is not accepted")
```
