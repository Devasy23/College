```python
import nltk
import gensim
nltk.download('punkt')
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
sentence_token = sent_tokenize(corpus)
```

```python
sentence_token
```

```python
story = []
for sentence in sentence_token:
  story.append(simple_preprocess(sentence))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window = 13,
    vector_size=30,
    min_count = 4
)
```

```python
model.build_vocab(story)
```

```python
print(model.epochs)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=20)
```

```python
model.wv['king']
```

```python
model.wv.similarity('daenerys','drogo')
```

```python
model.wv.most_similar('drogo')
```
