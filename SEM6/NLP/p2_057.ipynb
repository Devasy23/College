```python
# Tokenization
text = " Fresh Turn in UP's OBC Politics as Deputy CM Maurya Backs Demand for Caste Census.UP Will Play a Big Role in India Becoming a $5 Trillion Economy. No Alliance With Tipra Motha Until It Quits Greater Tipraland Demand. Delhi Mayor Elections on Monday After Two Failed Attempts. BJP Ratlam District Functionary Held in Drug Trafficking Case. Amit Shah to Address Two Election Rallies in Tripura on Monday. PM Modi to Inaugurate Helicopter Factory in Tumakuru. Owaisi on Crackdown on Child Marriages. Bhagwant Mann Govt Faces Heat Over Sacrilege Case as Kotkapura Firing Victims Intensify Protest. Even Parliament Has Acknowledged Delhi's Air Quality Has Improved"
tokens = word_tokenize(text)
print("Tokens:", tokens)
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
# fitting
vect.fit(tokens)
# examine 
vect.get_feature_names_out()

# Removing Stop Words
stop_words = set(stopwords.words("english"))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
print("Filtered Tokens:", filtered_tokens)

# Stemming
ps = PorterStemmer()
stemmed_tokens = [ps.stem(token) for token in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)

# Text Pre-processing
def preprocess_text(text):
  tokens = word_tokenize(text)
  tokens = [token.lower() for token in tokens if token not in string.punctuation]
  stop_words = set(stopwords.words("english"))
  filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
  ps = PorterStemmer()
  stemmed_tokens = [ps.stem(token) for token in filtered_tokens]
  return stemmed_tokens

# testing
preprocessed_text = preprocess_text("Fresh politics Deputy will Play after air big role trillion tripura turn two until becoming backs.")
print("Preprocessed Text:", preprocessed_text)
test_dtm = vect.transform(preprocessed_text)
#print(test_dtm)
df=pd.DataFrame(test_dtm.toarray(), columns=vect.get_feature_names())
df.head()
```
