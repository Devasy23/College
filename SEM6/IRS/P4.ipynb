```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import re
# Define the product review
review = "I bought this product and it was amazing. The quality was great and it exceeded my expectations."

# Tokenize the review
# tokens = # Define the product review
review = "I bought this product and it was amazing. The quality was great and it exceeded my expectations."
tokens = word_tokenize(review)

# Split the review into tokens based on white spaces and punctuation marks
# tokens = [token.strip() for token in re.findall(r'\b\w+\b|[.,;!?]', review)]

# Perform POS tagging
tagged_tokens = pos_tag(tokens)

# Extract features and opinions as described in the previous answer
# Extract features as frequent nouns
features = [token for token, pos in tagged_tokens if pos in ['NN', 'NNS', 'NNP', 'NNPS']]

# Extract opinions as words near frequent nouns
opinions = []
for i in range(len(tagged_tokens)):
    if tagged_tokens[i][1] in ['NN', 'NNS', 'NNP', 'NNPS']:
        if i > 0:
            opinions.append(tagged_tokens[i-1][0])
        if i < len(tagged_tokens)-1:
            opinions.append(tagged_tokens[i+1][0])

# Display the extracted features and opinions
print("Features:", features)
print("Opinions:", opinions)
```

```python
# perform Extract features and opinions from the product review to enhance the performance of the traditional recommender system. To extract features, consider frequent nouns and to extract opinions consider nearer words from a frequent noun. Use Part Of Speech tagging as a preprocessing technique before extracting features and opinions from product reviews.
# on kaggle product review dataset
import pandas as pd
df = pd.read_csv('train_p4.csv')
df.head()
 
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def preprocess(text):
    # Convert to lower case
    text = text.lower()
    # Remove punctuations
    text = re.sub(r'[^\w\s]', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Remove whitespaces
    text = text.strip()
    # remove stop words
    text = ' '.join([word for word in text.split() if word not in stopwords])
    
    return text

# Apply the preprocess function to the review text
df_text = df['reviews.text'].apply(preprocess)
# convet the review text into list
df_text = df_text.tolist()

 
 
```

```python
# now apply the pos tagging on the review text
# Extract features and opinions from the review text
# Define a function to extract features and opinions from the review text
def extract_features_opinions(text):
    # Tokenize the review
    tokens = word_tokenize(text)
    # Perform POS tagging
    tagged_tokens = pos_tag(tokens)
    # Extract features and opinions
    # map opinion to feature
    dict = {} # feature : opinion
    for i in range(len(tagged_tokens)):
        if tagged_tokens[i][1] in ['NN', 'NNS', 'NNP', 'NNPS']:
            if i > 0:
                dict[tagged_tokens[i][0]] = tagged_tokens[i-1][0]
            if i < len(tagged_tokens)-1:
                dict[tagged_tokens[i][0]] = tagged_tokens[i+1][0]
    return dict
    
extract_features_opinions(preprocess(review))
```

```python
# run it on df_text to make 2 columns one for features and other for opinions

df_text = df['reviews.text'].apply(preprocess)
# df_text = df_text.tolist()
df['features'] = df_text.apply(lambda x: extract_features_opinions(x)[0])
df['opinions'] = df_text.apply(lambda x: extract_features_opinions(x)[1])
df.head()
```

```python
df = df[['features', 'opinions', 'reviews.rating', 'reviews.text']]
df.to_csv('features_opinions.csv')
```
